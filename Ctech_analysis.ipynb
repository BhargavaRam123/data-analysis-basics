```python
import pandas as pd
df = pd.read_excel(r"C:\Users\bharg\Downloads\SRMIST CPS01-CTECH-Batches Configuration.xlsx")
df = df.sort_values("CGPA (Upto 3rd SEM)",ascending=False)
pd.set_option("display.max_rows", None) 
pd.set_option("display.max_columns", None) 
df.reset_index(drop=True,inplace = True)
df
```

```python
df['Name'] = df['Name'].str.replace('^[a-zA-Z_ ]*$',' ' )
df['Name'] = df['Name'].str.strip(".")
df
```

```python

df['Section']=df['Section'].fillna('z1')
def sortsections(x):
    # print("x value",x)
    ans = ''
    f=0
    for v in x:
        if(v.isalpha() and f==0):
            ans+=v.upper()
            f=1
        elif(v.isnumeric()):
            ans+=v
    return ans
df['Section'] = df['Section'].apply(lambda x:sortsections(x))
df.groupby('Section')['CGPA (Upto 3rd SEM)'].std().sort_values(ascending=True)
# df.groupby('Section').mean(numeric_only=True)
```

```python
df.groupby('Section')['CGPA (Upto 3rd SEM)'].mean().sort_values(ascending=True)
```
